```python
raw_data = pd.read_csv("raw_SDOH_data_utf8.csv", sep="\t")
```

```python
selected_data = raw_data[["Local Health District",
          "Locality",
          "Year of YearDate",
          "PopEst Total Population",
          "Education Spending per Student", 
          "Per Capita Income", 
          "Per Capita Income Maintenance Benefits", 
          "Per Job Income", 
          "Children in Poverty", 
          "Population in Poverty", 
          "Unemployment Rate", 
          "Uninsured Children", 
          "Uninsured Population2", 
          "Voter Turnout2",
          "Cost Burdened Households",
          "Home Ownership Rate",
          "Working Age",
          "Working Age Population w/ a Bachelor Degree",
          "Working Age Population w/o a HS Diploma",
          "Arrests per 1,000 Residents",
          ]]
```

```python
presidential_years = [2012, 2016]
senate_years = [2012, 2014]
house_years = [2010, 2012, 2014, 2016]
va_election_years = [2013]

selected_data.loc[:, 'Presidential Election'] = selected_data["Year of YearDate"].isin(presidential_years)
selected_data.loc[:, 'Senate Election'] = selected_data["Year of YearDate"].isin(senate_years)
selected_data.loc[:, 'House Election'] = selected_data["Year of YearDate"].isin(house_years)
selected_data.loc[:, 'VA Election'] = selected_data["Year of YearDate"].isin(va_election_years)
selected_data.loc[:,"Locality, Local Health District"] = selected_data["Locality"].astype(str) + ", " + selected_data["Local Health District"].astype(str)

selected_data = selected_data.drop(["Year of YearDate", "Locality", "Local Health District"], axis=1)
```

```python
selected_data.dtypes
```

```python
def convert_currency_to_numeric(df):
    for col in df.columns:
        if df[col].dtype == 'object':  # Check if column is string type
            if any('$' in str(x) for x in df[col]):  # Check if any value contains '$'
                df[col] = df[col].str.replace('$','', regex=False).str.replace(',', '', regex=False).astype(float)

    return df

def convert_percentage_to_numeric(df):
    for col in df.columns:
        if df[col].dtype == 'object':
            df[col] = df[col].str.rstrip('%').astype(float) / 100   
    return df
```

```python

cols_with_dollars = selected_data.loc[:,["Education Spending per Student",                  
"Per Capita Income",                                 
"Per Capita Income Maintenance Benefits",            
"Per Job Income",]]     

cols_with_dollars_converted = convert_currency_to_numeric(cols_with_dollars)

cols_with_percent = selected_data.loc[:,["Children in Poverty",                           
"Population in Poverty",                         
"Unemployment Rate",                                 
"Uninsured Children",                               
"Uninsured Population2",                             
"Voter Turnout2",                                   
"Cost Burdened Households",                          
"Home Ownership Rate",                               
"Working Age",                                       
"Working Age Population w/ a Bachelor Degree",       
"Working Age Population w/o a HS Diploma",]]     

cols_with_percent_converted = convert_percentage_to_numeric(cols_with_percent)

# cols_with_dollars_converted
cols_with_percent_converted
```

```python
concatenated_df = pd.concat([cols_with_dollars_converted, cols_with_percent_converted], axis=1)

cols_to_keep = selected_data.columns.difference(concatenated_df.columns)
cols_to_keep
other_data = selected_data[cols_to_keep]

reference_df = pd.concat([other_data, concatenated_df], axis=1)

reference_df.to_csv("reference_data.tsv", sep="\t")

reference_df
data = reference_df.drop("Locality, Local Health District", axis=1)
```

```python
data_features = data.drop("Voter Turnout2", axis=1)
data_labels = data["Voter Turnout2"].copy()
```

```python
#Prepare data for machine learning models:
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import FunctionTransformer

X_train, X_test, y_train, y_test = train_test_split(data_features, data_labels,  test_size=0.2)

num_attributes = X_train.select_dtypes(include=['number']).columns
bool_attributes = X_train.select_dtypes(exclude=['number']).columns

num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy="median")),
    ('std_scaler', StandardScaler()),
])

# Function to convert boolean to int
def convert_bool_to_int(X):
    return X.astype(int)

bool_pipeline = Pipeline([
    ('bool_to_int', FunctionTransformer(convert_bool_to_int, validate=False)),
])

# Combine pipelines into a full pipeline
full_pipeline = ColumnTransformer([
    ("num", num_pipeline, num_attributes),
    ("bool", bool_pipeline, bool_attributes),
])

X_train = full_pipeline.fit_transform(X_train)
X_test = full_pipeline.transform(X_test)
y_train = y_train
y_test = y_test
```
